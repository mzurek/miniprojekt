### 0. Data preprocessing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math

terror=pd.read_csv('globalterrorismdb_0718dist.csv',encoding='ISO-8859-1', low_memory=False)

terror.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region',
                       'success':'Success','attacktype1_txt':'AttackType','targtype1_txt':'TargetType',
                       'natlty1_txt':'TargetNationality','gname':'GroupName','weaptype1_txt':'WeaponType',
                       'nkill':'Killed','nwound':'Wounded'}, inplace=True)
terror=terror[['Year','Month','Day','Country','Region','Success','AttackType','TargetType','TargetNationality','GroupName','WeaponType','Killed','Wounded']]

terror['Casualties'] = terror['Killed']+terror['Wounded']
```

### 1. Terrorist attacks during certain seasons

```python
def getSeason(day,month):
    if month in [4,5] or (month==3 and day>=21) or (month==6 and day<21):
        return 'spring'
    elif month in [7,8] or (month==6 and day>=21) or (month==9 and day<23):
        return 'summer'
    elif month in [10,11] or (month==9 and day>=23) or (month==12 and day<22):
        return 'autumn'
    else:
        return 'winter'
        


indexes = list(set(terror['Month'].nonzero()[0]).intersection(terror['Day'].nonzero()[0]))
data = terror.iloc[indexes,[1,2]]

for i in indexes:
    data.at[i,'Season'] = getSeason(data.at[i,'Day'],data.at[i,'Month'])
    

values = [sum(data['Season']=='spring'),sum(data['Season']=='summer'),sum(data['Season']=='autumn'),
         sum(data['Season']=='winter')]
labels = ["spring","summer","autumn","winter"]

for i in range(0,4):
    print(labels[i],": ",values[i])
```

```python
fig1, ax1 = plt.subplots(figsize=(6, 5), subplot_kw=dict(aspect="equal"))
wedges, texts, autotexts  = ax1.pie(values, autopct='%1.2f%%',
        startangle=90)
ax1.axis('equal')

ax1.legend(wedges, labels,
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1),
          fontsize = 14)

plt.setp(autotexts, size=14)
plt.show()
```

### 2. The number of victims of terrorist attacks per region

```python
labels = np.array(terror['Region'].unique())
counts = np.zeros(len(labels))

for i in range(0,len(terror.index)):
    if(math.isnan(terror.at[i,'Casualties']) is not True):
        counts[np.where(labels==terror.at[i,'Region'])] += terror.at[i,'Casualties']
        
fig2, ax2 = plt.subplots()

ax2.barh(labels, counts,align='center',
        color='green', ecolor='black')

ax2.set_yticks(labels)
ax2.invert_yaxis() 

plt.show()
```

### 3. Type of attacks per year

```python
step = 5
i = 0

for year in terror['Year']:
    is_not_between = True
    start_year = 1970
    end_year = 1974
    while is_not_between:
        if(year<start_year or year>end_year):
            is_not_between = True
            start_year += step
            end_year += step
        else:
            is_not_between = False
            if end_year>2017: end_year = 2017
            terror.at[i,'TimePeriod'] = str(start_year)+"-"+str(end_year)
            i += 1

pd.crosstab(terror.TimePeriod,terror.AttackType).plot.barh(stacked=True,width=1,color=sns.color_palette('RdYlGn',9))
fig3=plt.gcf()
fig3.set_size_inches(12,8)
plt.show()
                
                    
            
```

### 4. Activity of 10 top terrorist groups

```python
top_groups_count = terror['GroupName'].value_counts()
top_groups_names = top_groups_count[1:11].index

top_groups = terror[terror['GroupName'].isin(top_groups_names)]
pd.crosstab(top_groups.Year,top_groups.GroupName).plot(color=sns.color_palette('Paired',10))
fig3=plt.gcf()
fig3.set_size_inches(18,6)
plt.autoscale(enable=True, axis='x', tight=True)
plt.show()
```

### 5. Number of terrorist attacks per year

```python
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=terror,palette='RdYlGn_r',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.show()
```
